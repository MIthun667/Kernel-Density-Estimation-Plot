```python
# the normal import
import numpy as np
from numpy.random import randn
import pandas as pd

# Import the stats Library from numpy
from scipy import stats

# these are ploting modules and Libraries we'll use
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = randn(25)
```

```python
sns.rugplot(dataset)
plt.ylim(0,1)
```

```python
plt.hist(dataset, alpha=0.5)
sns.rugplot(dataset)
```

```python
sns.rugplot(dataset)

x_min = dataset.min() - 2
x_max = dataset.max() + 2

x_axis = np.linspace(x_min, x_max, 100)
# we are actully using of estimation of the bandwidth equation(((4 * dataset.std()**5) / (3* len(dataset))) **0.2)
bandwidth = ((4 * dataset.std()**5) / (3* len(dataset))) **0.2

ker_list= []

for data_point in dataset: 
    #create a kernel for each point and append it to the kernel_lsit
    kernel = stats.norm(data_point, bandwidth).pdf(x_axis)
    ker_list.append(kernel)
    # Scale for plotting
    kernel = kernel / kernel.max()
    kernel = kernel *0.4
    plt.plot(x_axis, kernel, color= "gray", alpha = 0.5)
    
plt.ylim(0,1)
```

```python
sum_kde = np.sum(ker_list, axis=0)
fig = plt.plot(x_axis, sum_kde, color="red")
sns.rugplot(dataset)
plt.yticks([])
plt.suptitle("Sum of the basis function")
```
